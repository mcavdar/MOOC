```python
%%bash
tr -sc 'A-Za-z' '\n' < nyt_200811.txt > nyt.words
tail -n +2 nyt.words > nyt.nextwords
paste nyt.words nyt.nextwords > nyt.bigrams
tail -n +3 nyt.words > nyt.2nextwords
paste nyt.words nyt.nextwords nyt.2nextwords > nyt.trigrams
```

```python
%%bash
tr 'A-Z' 'a-z' < nyt.bigrams | sort | uniq -c | sort -nr | head -n 10
```

```python
%%bash
tr 'A-Z' 'a-z' < nyt.trigrams | sort | uniq -c | sort -nr | head -n 10
```
